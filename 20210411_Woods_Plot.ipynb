```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# read in the data

# df_apo = pd.read_csv('./uptake_summary_table_apo.csv', header=0) #change csv name
df_DNA = pd.read_csv('./uptake_summary_table_DNA.csv', header=0)
df_IPTG = pd.read_csv('./uptake_summary_table_IPTG.csv', header=0)
df_ONPF = pd.read_csv('./uptake_summary_table_ONPF.csv', header=0)
df_ONPFDNA = pd.read_csv('./uptake_summary_table_ONPFDNA.csv', header=0)
# df_TMG = pd.read_csv('./uptake_summary_table_TMG.csv', header=0)
```

```python
timepoints = df_DNA['Deut Time (sec)'].unique()
# peptides_apo = df_apo['Sequence'].unique()
peptides_DNA = df_DNA['Sequence'].unique()
peptides_IPTG = df_IPTG['Sequence'].unique()
peptides_ONPF = df_ONPF['Sequence'].unique()
peptides_ONPFDNA = df_ONPFDNA['Sequence'].unique()
# peptides_TMG = df_TMG['Sequence'].unique()
```

```python
mpl.rcParams.update({"figure.figsize": (15, 10 * len(timepoints))})
mpl.rcParams.update({"font.size": 26, "font.family": 'Arial'})
fig, ax2 = plt.subplots(len(timepoints)-1)

for i in range(0, len(timepoints)-1):

#     for j in range(len(peptides_apo)):
#         start = df_apo[df_apo['Sequence'] == peptides_apo[j]]['Start'].tolist()[0]
#         end = df_apo[df_apo['Sequence'] == peptides_apo[j]]['End'].tolist()[0]
#         xs = [start, end]
#         ys_apo = np.ones(2) * df_apo[(df_apo['Sequence'] == peptides_apo[j]) & (df_apo['Deut Time (sec)'] == timepoints[i])]['%D'].tolist()[0]
#         ax2[i-1].plot(xs, ys_apo, linewidth = 2, color='green')
    
    for j in range(len(peptides_DNA)):
        start = df_DNA[df_DNA['Sequence'] == peptides_DNA[j]]['Start'].tolist()[0]
        end = df_DNA[df_DNA['Sequence'] == peptides_DNA[j]]['End'].tolist()[0]
        xs = [start, end]
        ys_DNA = np.ones(2) * df_DNA[(df_DNA['Sequence'] == peptides_DNA[j]) & (df_DNA['Deut Time (sec)'] == timepoints[i])]['%D'].tolist()[0]
        ax2[i-1].plot(xs, ys_DNA, linewidth = 2, color='red')
        
    for j in range(len(peptides_IPTG)):
        start = df_IPTG[df_IPTG['Sequence'] == peptides_IPTG[j]]['Start'].tolist()[0]
        end = df_IPTG[df_IPTG['Sequence'] == peptides_IPTG[j]]['End'].tolist()[0]
        xs = [start, end]
        ys_IPTG = np.ones(2) * df_IPTG[(df_IPTG['Sequence'] == peptides_IPTG[j]) & (df_IPTG['Deut Time (sec)'] == timepoints[i])]['%D'].tolist()[0]
        ax2[i-1].plot(xs, ys_IPTG, linewidth = 2, color='blue')
    
    for j in range(len(peptides_ONPF)):
        start = df_ONPF[df_ONPF['Sequence'] == peptides_ONPF[j]]['Start'].tolist()[0]
        end = df_ONPF[df_ONPF['Sequence'] == peptides_ONPF[j]]['End'].tolist()[0]
        xs = [start, end]
        ys_ONPF = np.ones(2) * df_ONPF[(df_ONPF['Sequence'] == peptides_ONPF[j]) & (df_ONPF['Deut Time (sec)'] == timepoints[i])]['%D'].tolist()[0]
        ax2[i-1].plot(xs, ys_ONPF, linewidth = 2, color='orange')
            
    for j in range(len(peptides_ONPFDNA)):
        start = df_ONPFDNA[df_ONPFDNA['Sequence'] == peptides_ONPFDNA[j]]['Start'].tolist()[0]
        end = df_ONPFDNA[df_ONPFDNA['Sequence'] == peptides_ONPFDNA[j]]['End'].tolist()[0]
        xs = [start, end]
        ys_ONPFDNA = np.ones(2) * df_ONPFDNA[(df_ONPFDNA['Sequence'] == peptides_ONPFDNA[j]) & (df_ONPFDNA['Deut Time (sec)'] == timepoints[i])]['%D'].tolist()[0]
        ax2[i-1].plot(xs, ys_ONPFDNA, linewidth = 2, color='purple')
    
#     for j in range(len(peptides_TMG)):
#         start = df_TMG[df_TMG['Sequence'] == peptides_TMG[j]]['Start'].tolist()[0]
#         end = df_TMG[df_TMG['Sequence'] == peptides_TMG[j]]['End'].tolist()[0]
#         xs = [start, end]
#         ys_TMG = np.ones(2) * df_TMG[(df_TMG['Sequence'] == peptides_TMG[j]) & (df_TMG['Deut Time (sec)'] == timepoints[i])]['%D'].tolist()[0]
#         ax2[i-1].plot(xs, ys_TMG, linewidth = 2, color='gray')

    ax2[i-1].set_xlabel("Residue Number (N to C)")
    ax2[i-1].set_ylabel("% Deuteration")
    ax2[i-1].set_ylim([-5, 100])
    s = "t = " + str(timepoints[i]) + "s"
#     apo_label = "apo"
    DNA_label = "DNA"
    IPTG_label = "IPTG"
    ONPF_label = "ONPF"
    ONPFDNA_label = "ONPF-DNA"
#     TMG_label = "TMG"
    ax2[i-1].text(0, 90, s)
#     ax2[i-1].text(250, 85, apo_label, color="green")
    ax2[i-1].text(250, 90, DNA_label, color="red")
    ax2[i-1].text(250, 85, IPTG_label, color="blue")
    ax2[i-1].text(250, 80, ONPF_label, color="orange")
    ax2[i-1].text(250, 75, ONPFDNA_label, color="purple")
#     ax2[i-1].text(250, 60, TMG_label, color="gray")
plt.savefig("./WoodsPlot.png",  transparent=True)
```
