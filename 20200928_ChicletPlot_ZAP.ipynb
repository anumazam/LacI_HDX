```python
# Read in the data
df_apo = pd.read_csv('./ZAP_apo_2_092020.csv', header=0) #change csv name
df_bound = pd.read_csv('./ZAP_bound_2_092020.csv', header=0)
#df_ace2 = pd.read_csv('./20200807_ACE2_UptakeSummary.csv', header=0)
#Add column to reorganize data to be ordered by center of peptide not N-term
df_apo['Center'] = ((df_apo['End'] - df_apo['Start'])/2 + df_apo['Start']).astype('int32')
df_bound['Center'] = ((df_bound['End'] - df_bound['Start'])/2 + df_bound['Start']).astype('int32')
#df_ace2['Center'] = ((df_ace2['End'] - df_ace2['Start'])/2 + df_ace2['Start']).astype('int32')
#sort data based on the center column
df_apo = df_apo.sort_values(by=['Center'])
df_bound = df_bound.sort_values(by=['Center'])
#df_ace2 = df_ace2.sort_values(by=['Center'])
#Optional line (uncomment) to drop low confidence peptides
#df_apo.drop(df_apo[df_apo['Confidence'] == 'Low'].index, inplace = True)
#df_ab.drop(df_ab[df_ab['Confidence'] == 'Low'].index, inplace = True)
#df_ace2.drop(df_ace2[df_ace2['Confidence'] == 'Low'].index, inplace = True)

#Just a test to pull up some peptides
#df_apo[df_apo['Sequence'] == 'IYGKTVYH'].head()
```
